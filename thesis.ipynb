```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

waveform, sample_rate = librosa.load("./voice/fujitou_silence/fujitou_normal_001.wav")

mfcc = librosa.feature.mfcc(y = waveform,sr = sample_rate, n_mfcc = 13)
print(mfcc.shape)

librosa.display.specshow(mfcc, sr = sample_rate, x_axis = 'time', y_axis = 'frames')
plt.ylabel("MFCC")
plt.colorbar()
plt.show()
```
